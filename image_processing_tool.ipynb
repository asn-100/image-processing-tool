```python
# import packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# search for the first image in folder sample_images
image_folder = "sample_images"
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith((".jpg", ".jpeg", ".png"))]


if not image_files:
    raise FileNotFoundError("no picture found in sample_images-folder")

image_path = os.path.join(image_folder, image_files[0])

print(f"Load picture: {image_path}")
```

```python
# load pictures
img = cv2.imread(image_path)
```

```python
# convert to grey
gray_pic = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# recognize edges
edges_pic = cv2.Canny(gray_pic, 100, 200)
```

```python
# plot the results
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(gray_pic, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(edges_pic, cmap='gray')
plt.title("Edges (Canny)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
